#### Cleaning

```python
attributes_df  = pd.read_csv('01_catmemexp_attributestask_allmodels_processed_2.csv', dtype=str, quoting=csv.QUOTE_ALL, encoding = "utf-8")
attributes_df.head()
```

```python
attributes_df['output_list_processed'] = attributes_df['output_list'].apply(ast.literal_eval)
```

```python
attributes_df['output_list_processed'] = attributes_df['output_list'].copy()
```

```python
def process_outputs(row, transformations):
    processed_outputs = []
    for item in row['output_list_processed']: 
        processed_item = transformations.get(item, item)  
        if processed_item is not None:  
            processed_outputs.append(processed_item)
    return processed_outputs   

transformations = {
    "": None,
    " ": None}

attributes_df['output_list_processed'] = attributes_df.apply(lambda row: process_outputs(row, transformations), axis=1)
attributes_df.head()
```

```python
# remove None values
attributes_df.dropna(inplace=True)
```

```python
#removing duplicates
attributes_df['output_list_processed'] = attributes_df['output_list_processed'].apply(lambda x: list(set(x)))
```

```python

# Function to filter items based on frequency
def filter_items(group):

    group['output_list_processed'] = group['output_list_processed'].apply(
        lambda x: ast.literal_eval(x) if isinstance(x, str) else x
    )

    exploded = group['output_list_processed'].explode()
    
    attributes_count = exploded.value_counts()
    
    filtered_items = attributes_count[attributes_count > 1].index.tolist()

    group['output_list_processed_nonoise'] = group['output_list_processed'].apply(
        lambda x: [item for item in x if item in filtered_items]
    )
    
    return group

attributes_df = attributes_df.groupby(['model', 'category', 'item'], group_keys=False).apply(filter_items)

attributes_df.reset_index(drop=True, inplace=True)

attributes_df.head()
```

```python
attributes_df.head()
```

```python
attributes_df['output_list'] = attributes_df['output_list'].apply(ast.literal_eval)
```

```python
# Re-order output_list_processed
def reorder_column_processed(row):
    order_dict = {item: i for i, item in enumerate(row['output_list'])}
    return sorted(row['output_list_processed'], key=lambda x: order_dict.get(x, float('inf')))

attributes_df['output_list_processed'] = attributes_df.apply(reorder_column_processed, axis=1)

# Re-order output_list_processed_nonoise
def reorder_column_processedn(row):
    order_dict = {item: i for i, item in enumerate(row['output_list'])}
    return sorted(row['output_list_processed_nonoise'], key=lambda x: order_dict.get(x, float('inf')))

attributes_df['output_list_processed_nonoise'] = attributes_df.apply(reorder_column_processedn, axis=1)
```
