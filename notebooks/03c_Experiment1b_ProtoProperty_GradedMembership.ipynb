```python
# Setup
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import ast
import numpy as np
from collections import defaultdict
from matplotlib.ticker import MaxNLocator
```

```python
# Import Data

attributes_df  = pd.read_csv('01_catmemexp_attributestask_allmodels_processed.csv', dtype=str, quoting=csv.QUOTE_ALL, encoding = "utf-8")
attributes_df['output_list_processed_nonoise'] = attributes_df['output_list_processed_nonoise'].apply(ast.literal_eval)

items_df = pd.read_csv('01_catmemexp_itemstask_allmodels_processed.csv', dtype=str, quoting=csv.QUOTE_ALL, encoding="utf-8")
items_df['output_list_processed_nonoise'] = items_df['output_list_processed_nonoise'].apply(ast.literal_eval)


unique_attributes_df  = pd.read_csv('famres_scores_df.csv')
unique_attributes_df.head()
```

```python
unique_models = unique_attributes_df['model'].unique()

model_dataframes = {}

for model in unique_models:

    unique_categories = unique_attributes_df[unique_attributes_df['model'] == model]['category'].unique()

    common_attributes_counts = {}

    for category in unique_categories:

        filtered_df = unique_attributes_df[
            (unique_attributes_df['model'] == model) & 
            (unique_attributes_df['category'] == category)
        ]

        sorted_df = filtered_df.sort_values(by='frequency')

        most_frequent_items = sorted_df.tail(5)
        least_frequent_items = sorted_df.head(5)

        most_frequent_items['attributes'] = most_frequent_items['attributes'].apply(eval)
        least_frequent_items['attributes'] = least_frequent_items['attributes'].apply(eval)

        most_common_attributes = set(most_frequent_items['attributes'].iloc[0])  
        for attrs in most_frequent_items['attributes'].iloc[1:]:
            most_common_attributes.intersection_update(attrs)

        most_common_count = len(most_common_attributes)

        least_common_attributes = set(least_frequent_items['attributes'].iloc[0])  
        for attrs in least_frequent_items['attributes'].iloc[1:]:
            least_common_attributes.intersection_update(attrs)

        least_common_count = len(least_common_attributes)

        common_attributes_counts[category] = {
            'Category': category,
            'Model': model,
            'Most typical members': most_common_count,
            'Least typical members': least_common_count,
        }

    model_summary_df = pd.DataFrame.from_dict(common_attributes_counts, orient='index')

    model_dataframes[model] = model_summary_df

concatenated_df = pd.concat(model_dataframes.values(), ignore_index=True)

concatenated_df.to_csv('combined_freq_attr_summary.csv', index=False)

print("\nCombined Summary DataFrame:")
concatenated_df
```
